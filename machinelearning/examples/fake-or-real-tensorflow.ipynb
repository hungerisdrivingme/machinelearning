```python
y=df["Type"].values
x=df.drop("Type", axis=1)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train= scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
 
```

```python
model=Sequential()
```

```python
model.add(Dense(units=30, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="Adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
loss=model.history.history
loss=pd.DataFrame(loss)
```

```python
loss
```

```python
loss.plot()
```

# earlystopping

```python
model2=Sequential()

model2.add(Dense(units=30, activation="relu"))
model2.add(Dense(units=15, activation="relu"))
model2.add(Dense(units=15, activation="relu"))
model2.add(Dense(units=1, activation="sigmoid"))

model2.compile(loss="binary_crossentropy",optimizer="Adam")
```

```python
es=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model2.fit(x=x_train, y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[es])
```

```python
loss2=pd.DataFrame(model2.history.history)
```

```python
loss2.plot()
```

# dropout

```python
model3=Sequential()

model3.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.4))
model3.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.4))
model3.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.4))
model3.add(Dense(units=1, activation="sigmoid"))

model3.compile(loss="binary_crossentropy",optimizer="Adam")
```

```python
model3.fit(x=x_train, y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[es])
```

```python
loss3=pd.DataFrame(model3.history.history)
```

```python
loss3.plot()
```
