```python
df_new=df_new[df_new.year != 1970]
```

```python
df_new=df_new.drop("transmission",axis=1)
```

```python
df_new
```

```python
y=df_new["price"].values
x=df_new.drop("price",axis=1).values
type(y),type(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(
x, y, test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
x_train=scale.fit_transform(x_train)
x_test=scale.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="Adam",loss="mse")
```

```python
model.fit(x=x_train, y=y_train ,validation_data=(x_test,y_test), epochs=150, batch_size= 250)
```

```python
loss=pd.DataFrame(model.history.history)
loss
```

```python
loss.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_squared_log_error
```

```python
pred=model.predict(x_test)
```

```python
plt.scatter(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
mean_squared_log_error(y_test,pred)
```
