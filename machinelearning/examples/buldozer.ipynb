```python
def add_datepart(df2,dt_name,drop=True,time=False): #creates new columns from our dt column
                                                    #optimal spliti yapabilmesi icin hangi gÃ¼nlerin haftasonu
                                                    #oldugunu sisteme Ã¶gretmek icin ayÄ±rÄ±yoruz.
    dt_column=df2[dt_name]
    column_dtype=dt_column.dtype
    targ_name=re.sub('[Dd]ate$','',dt_name)
    
    attr=['Year','Month','Week','Day','Dayofweek','Dayofyear','Is_month_end',
         'Is_month_start','Is_quarter_end','Is_quarter_start','Is_year_end','Is_year_start']
    
    if time:
        attr=attr+['Hour','Minute','Second']
        
    for a in attr:
        df2[targ_name+a]=getattr(dt_column.dt,a.lower())
    #howmuch time passed,we will divide by 10^9 because it is in the nanosec format   
    df2[targ_name+'Elapsed']=dt_column.astype(np.int64)//10**9
    
    if drop:
        df2.drop(dt_name,axis=1,inplace=True)
        
```

```python
#train kÄ±smÄ±
def train_cats(df2):
    for n,c in df2.items():  #n column name,  c series,, Tuple donduruyor ilk elemanÄ± n ikinci el c
        if is_string_dtype(c):  #string dtypeÄ± pandas kategoriye cevirir
            df[n]=c.astype("category").cat.as_ordered()
```

```python
#traine train_cats uyguladÄ±ktan sonra aynÄ± degiÅŸimleri validation ve teste de uygulamalÄ±yÄ±z
#validation kÄ±smÄ±    #amac traine yapÄ±lan kategorik islemin aynÄ±sÄ±nÄ±n test ve valid'e de uygulanmasÄ±
def apply_cats(df2,train):
    for n,c in df2.items():
        if train[n].dtype=="category":
            df2[n]=pd.Categorical(c,categories=train[n].cat.categories,ordered= True)
```

```python
#Missing of the entry can also be a valuable information
#So we will create a column that is False when value is missing
#We encoded missingness in categorical columns so we will just create _na columns for numerical types
def fix_missing(df2,col,name):
    if is_numeric_dtype(col): #numerik mi?
        if pd.isnull(col).sum: #null var mÄ±? 0 false 1 true
            df2[name+"_na"]=pd.isnull(col)  #column yarattÄ±k
        df2[name]=col.fillna(col.median())   #missing valuelarÄ± medyan ile degistirdi
```

```python
def numericalize(df2,col,name):
    if not is_numeric_dtype(col):
        df2[name]=col.astype('category').values.codes+1      # missing valuelar -1 degil 0 olsun diye
```

```python
def proc_df(df2,y_fld):
    y=df2[y_fld].values
    df2.drop([y_fld],axis=1,inplace=True)
    for n,c in df2.items():
        fix_missing(df2,c,n)
    for n,c in df2.items():
        numericalize(df2,c,n)
    res=[df2,y]
    
    return res
```
